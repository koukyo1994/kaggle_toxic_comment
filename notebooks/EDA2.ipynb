```python
import wordbatch
from wordbatch.models import FM_FTRL, FTRL
from nltk.corpus import stopwords
from wordbatch.extractors import WordBag
import re
```

```python
stop = {x: 1 for x in stopwords.words('english')}
non_alphanums = re.compile(u'[^A-Za-z0-9]+')


def normalize_text(text):
    return u" ".join(
        [x for x in [y for y in non_alphanums.sub(' ', text).lower().strip().split(" ")] \
         if len(x) > 1 and x not in stop])
```

```python
wb = wordbatch.WordBatch(normalize_text, extractor=(WordBag, {"hash_ngrams": 2, "hash_ngrams_weights": [1.5, 1.0],
                                                                  "hash_size": 2 ** 22, "norm": None, "tf": 'binary',
                                                                  "idf": None,
                                                                  }), procs=8)
```

```python
train_length = df_train.shape[0]
concat_text = pd.concat([df_train['comment_text'], df_test['comment_text']])
X_text = wb.fit_transform(concat_text)
```

```python
X_train = X_text[:train_length]
X_test = X_text[train_length:]
```

```python
from sklearn.cross_validation import train_test_split
import lightgbm as lgbm
X_train, X_valid, y_train, y_valid = train_test_split(X_train, df_train.iloc[:, 2:], test_size=0.1, random_state=13)
```

```python
lgb_train = lgbm.Dataset(X_train, y_train.iloc[:, 0])
lgb_eval = lgbm.Dataset(X_valid, y_valid.iloc[:, 0], reference=lgb_train)
```
